#### 1.Выберите наборы данных. которые можно интерпретировать, как временные ряды:

а) зависимость цены на недвижимость в конкретном регионе от времени,

б) скорость полета самолета при перелете из аэропорта А в аэропорт В,

в) количество отказов компьютера,

г) число обращений в поликлинику по району,

д) координаты автобуса при движении по маршруту,

е) координаты фигуры нарушителя в файле видеофиксации охранной системы.


Ответы:   
а) можно, цена зависит от множества параметров, в том числе и региона, имеет устоявшиеся зависимости цены от параметров объекта недвижимости. Например в двух соседних домах при сопостовимых параметрах не может быть большой разницы в цене и она не может резко измениться.

б) можно, скорость не может мгновенно измениться, определяется ускорением, т.е. скоростью изменения предыдущего значения скорости полёта.

в) нет, отказы не зависят от момента времени,когда их регистрируют. Отказы зависят от множества параметров, не зависящих друг от друга (например устойчивость ОС и стабильность электропитания).

г) нет, заболевание человека зависит от множества независимых обстоятельств.

д) можно, автобус не может мгновенно переместиться из одной точки в другую

е) можно, нарушитель не может мгновенно переститься в пространстве и соответственно будет появляться в камерах последовательно.

#### 3.Прочитайте базу Brent и сделайте ее перерасчет (ресемплинг) в представление по неделям , по месяцам.

```python
brent_data = pd.read_excel('./Data/нефть-brent 2010-2019.xlsx', index_col=[0], parse_dates=[0])
brent_data.head()
```

```python
_ = brent_data['Значение'].plot( style='.', figsize=( 15, 7 ), title='Brent ')
plt.ylabel('USA dollar')
plt.xlabel('t')
```

```python
brent_data_w = brent_data.resample('W').mean()
brent_data_w.head()
```

```python
brent_data_m = brent_data.resample('M').mean()
brent_data_m.head()
```

#### 4.Постройте график изменения среднего по выборке и среднего отклонения по выборке на недельных, месячных и годовых группах для Brent.

##### Графики средних значений

```python
brent_data_w_m = brent_data.resample('W').mean()
brent_data_w_m['Значение'].plot( style='.', figsize=( 15, 7 ), title='Brent ')
plt.ylabel('USA dollar')
plt.xlabel('t')
```

```python
brent_data_m_m = brent_data.resample('M').mean()
brent_data_m_m['Значение'].plot( style='.', figsize=( 15, 7 ), title='Brent ')
plt.ylabel('USA dollar')
plt.xlabel('t')
```

```python
brent_data_y_m = brent_data.resample('Y').mean()
brent_data_y_m['Значение'].plot( style='.', figsize=( 15, 7 ), title='Brent ')
plt.ylabel('USA dollar')
plt.xlabel('t')
```

##### Графики средних стандартных отклонений

```python
brent_data_w_m = brent_data.resample('W').std()
brent_data_w_m['Значение'].plot( style='r', figsize=( 15, 7 ), title='Brent ')
plt.ylabel('USA dollar')
plt.xlabel('t')
```

```python
brent_data_m_m = brent_data.resample('M').std()
brent_data_m_m['Значение'].plot( style='r', figsize=( 15, 7 ), title='Brent ')
plt.ylabel('USA dollar')
plt.xlabel('t')
```

```python
brent_data_y_m = brent_data.resample('Y').std()
brent_data_y_m['Значение'].plot( style='r', figsize=( 15, 7 ), title='Brent ')
plt.ylabel('USA dollar')
plt.xlabel('t')
```
